```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
clust_df = pd.read_csv("/content/sample_data/iris.data")
clust_df.head()
```

```
col_name = [col for col in clust_df.columns]
col_name
```

```
df = clust_df.drop('class',axis = 1)
df.head()
```

```
from sklearn.preprocessing import StandardScaler
X = df.values[:,:]
X
```

```
x = np.nan_to_num(X)
x

```

```
Clus_dataSet = StandardScaler().fit_transform(x)
Clus_dataSet
```

```
from sklearn.cluster import KMeans
error = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(Clus_dataSet)
    error.append(kmeans.inertia_)
```

```
plt.plot(range(1, 11), error)
plt.title('Result')
plt.xlabel('Number of clusters')
plt.ylabel('Error')
plt.show()
```

```
kmeans = KMeans(n_clusters = 3)
y_kmeans = kmeans.fit_predict(x)
```

```
for i in range(0, 3):
    for j in range(i+1, 4):
        ax = plt.subplot()
        plt.xlabel(col_name[i])
        plt.ylabel(col_name[j])
        plt.scatter(x[y_kmeans == 0, i], x[y_kmeans == 0, j], s = 100, c = 'green', label = 'Iris-setosa')
        plt.scatter(x[y_kmeans == 1, i], x[y_kmeans == 1, j], s = 100, c = 'red', label = 'Iris-versicolour')
        plt.scatter(x[y_kmeans == 2, i], x[y_kmeans == 2, j], s = 100, c = 'blue', label = 'Iris-virginica')
        plt.scatter(kmeans.cluster_centers_[:, i], kmeans.cluster_centers_[:,j], s = 100, c = 'yellow', edgecolors = 'black', label = 'Centroids')

        plt.legend()
        plt.show()
```
