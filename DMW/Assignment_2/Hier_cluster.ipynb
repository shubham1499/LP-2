```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```
clust_df = pd.read_csv("/content/sample_data/iris.data")
clust_df.head()
```

```
col_name = [col for col in clust_df.columns]
col_name
```

```
df = clust_df.drop('class',axis = 1)
df.head()
```

```
from sklearn.preprocessing import StandardScaler
X = df.values[:,:]
x = np.nan_to_num(X)
```

```
from scipy.cluster.hierarchy import dendrogram, linkage
links = linkage(x, 'single')
plt.figure(figsize=(8, 6))
dendrogram(links, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.show()

```

```
from sklearn.cluster import AgglomerativeClustering
y = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
y.fit_predict(x)
```

```
for i in range(0, 3):
    for j in range(i+1, 4):
        plt.subplot()
        plt.xlabel(col_name[i])
        plt.ylabel(col_name[j])
        plt.scatter(x[:,i], x[:,j], s = 100, c = y.labels_, cmap='brg')
        plt.show()
```
